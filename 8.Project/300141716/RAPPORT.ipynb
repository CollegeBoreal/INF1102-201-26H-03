## 🎯 Objectif du projet

Ce projet a pour objectif d’analyser le prix de l’essence à Toronto

L’objectif est de comprendre comment les événements au Moyen-Orient 
peuvent influencer les prix locaux.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 📂 Chargement des données

On charge un fichier CSV contenant :
- les dates
- le prix de l’essence à Toronto
- le prix du pétrole Brent

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/prix_energie.csv")
df
```

## 📊 Statistiques descriptives

On affiche les statistiques principales :
- moyenne
- minimum
- maximum

```python
df.describe()
```

```python
print("Prix moyen essence :", round(df["essence_toronto"].mean(), 2), "$/L")
print("Prix minimum essence :", df["essence_toronto"].min(), "$/L")
print("Prix maximum essence :", df["essence_toronto"].max(), "$/L")

print("Prix moyen Brent :", round(df["brent"].mean(), 2), "$/baril")
print("Prix minimum Brent :", df["brent"].min(), "$/baril")
print("Prix maximum Brent :", df["brent"].max(), "$/baril")
```

## 📈 Évolution du prix de l’essence

```python
plt.figure(figsize=(8,4))
plt.plot(df["date"], df["essence_toronto"], marker="o")
plt.title("Évolution du prix de l'essence à Toronto")
plt.xlabel("Date")
plt.ylabel("Prix ($/L)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 📈 Évolution du prix du Brent

```python
plt.figure(figsize=(8,4))
plt.plot(df["date"], df["brent"], marker="o")
plt.title("Évolution du prix du Brent")
plt.xlabel("Date")
plt.ylabel("Prix ($/baril)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 📊 Analyse

On remarque que :

- le prix de l’essence augmente progressivement
- le prix du Brent augmente aussi
- les deux courbes évoluent dans le même sens

Cela montre qu’une hausse du pétrole brut peut influencer les prix de l’essence à Toronto.
