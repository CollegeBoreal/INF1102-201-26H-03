The most frequent keywords reveal the dominant concerns in the dataset.

The project classifies articles into major cyber domains such as Data Breach, Ransomware, Phishing, Cloud Security, and Identity & Access.

Based on the generated report, the dominant theme is Identity & Access. This shows how critical authentication, access control, and identity governance have become in modern cybersecurity environments.

## 5. Severity Analysis

The articles are also evaluated according to severity.

Three levels are used:
- Low
- Medium
- High

The report shows that most articles fall into the Medium category, while a smaller number are considered High severity. This suggests that the dataset contains both strategic discussion and real operational risk signals.

## 6. Cyber Threat Index (CTI)

The Cyber Threat Index is a custom score created for this project.

It is based on weighted threat vocabulary:
- high-risk terms have the strongest weight
- medium-risk terms have moderate weight
- low-risk terms have lower weight

This makes it possible to rank articles by cyber-risk intensity and identify the most critical ones quickly.

```python
import json

with open("data/news_data.json", "r", encoding="utf-8-sig") as f:
    articles = json.load(f)

len(articles)

import re
from collections import Counter

def clean(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    return text

full_text = " ".join(a["title"] + " " + a["summary"] for a in articles)
words = clean(full_text).split()

Counter(words).most_common(10)

import matplotlib.pyplot as plt

top = Counter(words).most_common(10)

labels = [w for w, c in top]
values = [c for w, c in top]

plt.figure(figsize=(10, 5))
plt.bar(labels, values)
plt.xticks(rotation=30)
plt.title("Top Cyber Keywords")
plt.show()

```
