## 🎯 Objectif

Ce projet a pour objectif de mettre en œuvre une approche **Infrastructure as Code (IaC)** afin de déployer automatiquement une machine virtuelle Ubuntu, puis d’y configurer un serveur web NGINX accessible via une adresse IP.

L’objectif pédagogique est de comprendre comment décrire une infrastructure sous forme de code, automatiser le déploiement d’une VM, établir une connexion sécurisée en SSH et rendre un site web accessible à distance.


## 🏗️ Environnement technique

- Hyperviseur : **Proxmox VE**  
- Outil IaC : **OpenTofu**  
- Système d’exploitation : **Ubuntu Server**  
- Serveur Web : **NGINX**  
- Accès distant : **SSH**  
- Adresse IP de la VM : **10.7.237.210**

Le projet a été réalisé sur une machine virtuelle Ubuntu déployée automatiquement à partir d’un template Proxmox, puis configurée pour héberger une page web personnalisée.


## ⚙️ Étapes du projet

### 1. Déploiement de la machine virtuelle

La machine virtuelle Ubuntu a été créée automatiquement à l’aide d’OpenTofu. Cette étape a permis de définir la VM sous forme déclarative avec les paramètres réseau, mémoire, processeur et disque.

### 2. Connexion SSH

Une fois la machine déployée, la connexion distante a été réalisée en SSH à partir du poste client afin d’administrer le serveur Linux.

### 3. Installation de NGINX

Le serveur web NGINX a été installé sur Ubuntu à l’aide du gestionnaire de paquets `apt`.

### 4. Déploiement du site web

Une page HTML/CSS a été créée pour présenter le projet sous forme de portfolio. Cette page a ensuite été copiée dans le répertoire `/var/www/html` afin d’être servie par NGINX.

### 5. Vérification

Le bon fonctionnement a été vérifié avec le navigateur ainsi qu’avec la commande `curl` afin de confirmer que le serveur répond correctement.


## 💻 Commandes principales utilisées

### Déploiement IaC

```bash
tofu init
tofu plan
tofu apply
```

### Connexion SSH

```bash
ssh -i ~/.ssh/id_ed25519 ubuntu@10.7.237.210
```

### Installation NGINX

```bash
sudo apt update
sudo apt install nginx -y
```

### Vérification du service

```bash
sudo systemctl status nginx
```

### Vérification du site

```bash
curl http://localhost
curl http://10.7.237.210
```
