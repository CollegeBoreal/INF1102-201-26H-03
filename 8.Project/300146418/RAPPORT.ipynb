## 🎯 Introduction approfondie

Dans le domaine des systèmes informatiques modernes, les méthodes traditionnelles de configuration manuelle des infrastructures présentent plusieurs limites importantes, notamment en termes de temps, de fiabilité et de reproductibilité. En effet, la configuration manuelle est souvent sujette à des erreurs humaines et rend difficile la reproduction exacte d’un environnement sur plusieurs machines.

Pour répondre à ces défis, une approche appelée **Infrastructure as Code (IaC)** a été développée. Cette méthode consiste à décrire l’infrastructure sous forme de code, permettant ainsi d’automatiser entièrement son déploiement. Cela permet non seulement de gagner du temps, mais aussi d’assurer une cohérence et une standardisation des environnements.

Dans ce projet, cette approche a été mise en œuvre à l’aide de l’outil **OpenTofu**, afin de déployer automatiquement une machine virtuelle Ubuntu sur la plateforme Proxmox. Une fois la machine créée, un serveur web NGINX a été installé et configuré afin d’héberger un site web accessible via une adresse IP.

Ce projet constitue une première expérience concrète dans le domaine du **DevOps**, où l’automatisation et la gestion des infrastructures jouent un rôle essentiel.

## 💻 Étapes techniques détaillées

### 🔹 1. Déploiement de la machine virtuelle

Les commandes suivantes ont été utilisées pour déployer la machine virtuelle :

```bash
tofu init
tofu plan
tofu apply
```

| Commande | Fonction |
|---------|---------|
| `tofu init` | Initialise le projet et les providers |
| `tofu plan` | Affiche les changements prévus |
| `tofu apply` | Lance réellement le déploiement |

Cette étape est fondamentale car elle transforme l’infrastructure en code exécutable. Plutôt que de créer manuellement la machine virtuelle depuis l’interface graphique, OpenTofu permet de reproduire le même environnement avec les mêmes paramètres aussi souvent que nécessaire.


### 🔹 2. Connexion SSH

Une fois la machine virtuelle créée, la connexion distante a été établie avec SSH :

```bash
ssh ubuntu@10.7.237.210
```

SSH a permis d’administrer le serveur à distance, ce qui est une pratique standard dans les environnements Linux. Cette étape confirme également que le réseau de la machine virtuelle a été correctement configuré et que le serveur répond comme prévu.


### 🔹 3. Mise à jour du système

Avant l’installation des services, le système a été mis à jour :

```bash
sudo apt update
sudo apt upgrade -y
```

| Action | Impact |
|-------|-------|
| `apt update` | Récupère la liste des paquets disponibles |
| `apt upgrade -y` | Installe les mises à jour |

Cette étape permet d’éviter les problèmes de sécurité ou de compatibilité. Dans un contexte professionnel, une machine non mise à jour représente un risque important.


### 🔹 4. Installation de NGINX

Le serveur web NGINX a ensuite été installé :

```bash
sudo apt install nginx -y
```

Vérification du service :

```bash
sudo systemctl status nginx
```

Résultat attendu : **active (running)**

NGINX a été choisi parce qu’il est léger, performant et très largement utilisé dans les environnements réels. Son installation et sa configuration permettent de transformer la machine virtuelle en serveur web pleinement fonctionnel.


### 🔹 5. Déploiement du site web

Le site web a été créé sous forme d’un portfolio HTML/CSS, puis copié dans le répertoire de publication de NGINX.

```bash
sudo nano /var/www/html/index.html
```

Vérification de la présence du fichier :

```bash
ls /var/www/html
```

Cette étape a permis de publier une interface web personnalisée afin de démontrer que le serveur était non seulement en ligne, mais aussi capable de servir un contenu concret et structuré.


### 🔹 6. Vérification du fonctionnement

Le bon fonctionnement du serveur web a été validé avec plusieurs méthodes :

```bash
curl http://localhost
curl http://10.7.237.210
```

En plus du terminal, un test dans le navigateur a permis de confirmer que le site était bien accessible à l’adresse IP publique attribuée à la machine virtuelle.

Ces vérifications sont essentielles car elles permettent de valider la chaîne complète : système → service web → réseau → navigateur.

## 🧠 Analyse technique approfondie

L’Infrastructure as Code apporte une réelle transformation dans la manière de gérer les environnements serveurs. Dans une approche manuelle, chaque étape doit être reproduite à la main, ce qui prend du temps et augmente le risque d’oublier certains paramètres. Avec OpenTofu, la configuration devient déclarative : l’infrastructure souhaitée est décrite dans des fichiers, puis appliquée automatiquement.

Cette logique est particulièrement utile dans les environnements où plusieurs machines doivent être créées ou maintenues de manière cohérente. Elle facilite aussi la maintenance, car le code devient une documentation technique vivante.

Le choix de NGINX est également pertinent. En plus d’être simple à installer, ce serveur web est reconnu pour sa performance et sa stabilité. Il permet de publier rapidement un site statique et de vérifier le bon fonctionnement de la machine virtuelle déployée.
