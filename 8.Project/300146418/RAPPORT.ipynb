## ⚙️ Méthodologie

Le projet a été réalisé en plusieurs étapes :

1. Déploiement de la VM
2. Connexion SSH
3. Mise à jour du système
4. Installation NGINX
5. Déploiement du site
6. Vérification


## 💻 Étapes détaillées

### 1. Déploiement de la machine virtuelle

```bash
tofu init
tofu plan
tofu apply
```

Ces commandes permettent de créer automatiquement l’infrastructure.


### 2. Connexion SSH

```bash
ssh ubuntu@10.7.237.210
```

Permet d’accéder au serveur à distance.


### 3. Mise à jour du système

```bash
sudo apt update
sudo apt upgrade -y
```

Permet de maintenir le système à jour.


### 4. Installation NGINX

```bash
sudo apt install nginx -y
sudo systemctl status nginx
```

Le service doit être actif.


### 5. Déploiement du site web

```bash
sudo nano /var/www/html/index.html
```

Création du site web.


### 6. Vérification

```bash
curl http://localhost
```

Le site est accessible via navigateur.
