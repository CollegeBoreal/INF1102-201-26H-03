## 🎯 Objectif du projet

Ce projet consiste à déployer automatiquement une machine virtuelle Ubuntu en utilisant une approche Infrastructure as Code (IaC), puis à configurer un serveur web NGINX afin d’héberger un site web accessible via une adresse IP.

L’objectif est de comprendre comment automatiser le déploiement d’une infrastructure et configurer un service web fonctionnel.

## 🏗️ Environnement technique

- Proxmox VE (virtualisation)
- OpenTofu (Infrastructure as Code)
- Ubuntu Server
- NGINX
- SSH

Adresse IP de la machine : http://10.7.237.210


## ⚙️ Étapes réalisées

### 1. Déploiement de la machine virtuelle

La machine virtuelle a été créée automatiquement avec OpenTofu.

Commandes utilisées :
```bash
tofu init
tofu plan
tofu apply
```

Cette étape permet de créer une infrastructure reproductible et automatisée.


### 2. Connexion à la machine via SSH

Connexion à la machine virtuelle avec SSH.

```bash
ssh ubuntu@10.7.237.210
```

Cette étape permet d’accéder au serveur pour effectuer les configurations.


### 3. Mise à jour du système

```bash
sudo apt update
sudo apt upgrade -y
```

Permet d’assurer que tous les paquets sont à jour.


### 4. Installation du serveur web NGINX

```bash
sudo apt install nginx -y
```

Vérification du service :

```bash
sudo systemctl status nginx
```

Le service doit être actif (running).


### 5. Déploiement du site web

Création du fichier HTML :

```bash
sudo nano /var/www/html/index.html
```

Ajout du contenu HTML (portfolio).

Vérification :

```bash
ls /var/www/html
```


### 6. Vérification du site web

Test avec curl :

```bash
curl http://localhost
```

Test avec navigateur :

http://10.7.237.210

Le site doit s’afficher correctement.
