## Description

Ce projet consiste à analyser les actualités du site BBC afin d’identifier les mots les plus fréquents dans les titres.

```python
import json

with open("data/news_data.json", "r") as f:
    data = json.load(f)

data
```

```python
import pandas as pd

df = pd.DataFrame(list(data.items()), columns=["Mot", "Fréquence"])
df
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Mot"], df["Fréquence"])
plt.title("Mots les plus fréquents")
plt.xticks(rotation=45)
plt.show()
```

## Analyse des résultats

Les résultats montrent les mots les plus fréquents dans les titres des actualités du site BBC.

On observe que certains mots apparaissent avec la même fréquence, ce qui signifie qu’il n’y a pas de mot dominant dans cet échantillon.

Cela peut s’expliquer par le fait que peu d’articles ont été analysés.
