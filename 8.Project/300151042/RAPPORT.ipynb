## 🎯 Objectif
Ce notebook analyse un fichier de logs web et affiche les IP les plus actives les URLs les plus visitées et les codes HTTP.

```python
from collections import Counter
from datetime import datetime

file = "data/sample.log"

ips = []
urls = []
codes = []

with open(file, "r") as f:
    for line in f:
        parts = line.split()
        if len(parts) > 7:
            ips.append(parts[0])
            urls.append(parts[5])
            codes.append(parts[7])

top_ips = Counter(ips).most_common(5)
top_urls = Counter(urls).most_common(5)
top_codes = Counter(codes).most_common()

print("Fichier charge")
print("Total requetes:", len(ips))
```

```python
print("=" * 40)
print("RAPPORT ANALYSE LOGS")
print("=" * 40)
print("Date:", datetime.now())

print("\nTop IPs")
for ip, c in top_ips:
    print(ip, c)

print("\nTop URLs")
for url, c in top_urls:
    print(url, c)

print("\nCodes HTTP")
for code, c in top_codes:
    print(code, c)
```

```python
import matplotlib.pyplot as plt

labels = [ip for ip, _ in top_ips]
values = [c for _, c in top_ips]

plt.figure()
plt.bar(labels, values)
plt.title("Top IPs")
plt.xlabel("IP")
plt.ylabel("Requetes")
plt.show()
```

```python
now = datetime.now()

with open("output/rapport.txt", "w") as f:
    f.write("RAPPORT LOGS\n")
    f.write(str(now) + "\n")

    f.write("\nTop IPs\n")
    for ip, c in top_ips:
        f.write(f"{ip} {c}\n")

    f.write("\nTop URLs\n")
    for url, c in top_urls:
        f.write(f"{url} {c}\n")

    f.write("\nCodes HTTP\n")
    for code, c in top_codes:
        f.write(f"{code} {c}\n")

print("rapport.txt genere")
```
