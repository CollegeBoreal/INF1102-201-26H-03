# Rapport – Monitoring de site web

## 🎯 Objectif
Ce rapport analyse la disponibilité et le temps de réponse du site https://www.google.com

## 👤 Auteur

- ID Boréal : 300141625

```python
import requests
import time

url = 'https://www.google.com'
resultats = []

for i in range(5):
    debut = time.time()
    r = requests.get(url)
    duree = round(time.time() - debut, 2)
    resultats.append({'status': r.status_code, 'duree': duree})
    time.sleep(1)

print(resultats)
```

```python
import matplotlib.pyplot as plt

durees = [r['duree'] for r in resultats]

plt.figure(figsize=(8,4))
plt.plot(durees, marker='o', color='blue', label='Temps de réponse (s)')
plt.title('Monitoring - https://www.google.com')
plt.xlabel('Requête')
plt.ylabel('Secondes')
plt.legend()
plt.grid(True)
plt.savefig('output/graphique.png')
plt.show()
print('✅ Graphique généré')
```
