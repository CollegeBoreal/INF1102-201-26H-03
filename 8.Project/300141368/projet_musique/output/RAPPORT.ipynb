```python
import json
import matplotlib.pyplot as plt
```

```python
with open("../output/data.json", encoding="utf-8") as f:
    data = json.load(f)

artists = data['artists']['artist'][:5]
names = [a['name'] for a in artists]
plays = [int(a['playcount']) for a in artists]
```

```python
for i in range(len(names)):
    print(f"{names[i]} : {plays[i]} écoutes")
```

```python
plt.bar(names, plays)
plt.xticks(rotation=30)
plt.title("Top artistes Last.fm")
plt.show()
```
