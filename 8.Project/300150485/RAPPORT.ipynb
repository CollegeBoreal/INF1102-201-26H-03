# Analyse de Logs - Projet 8

Analyse des IP et URLs les plus fréquentes.

```python
from collections import Counter

file = "data/sample.log"

ips = []
urls = []

with open(file) as f:
    for line in f:
        parts = line.split()
        if len(parts) >= 6:
            ips.append(parts[0])
            urls.append(parts[5].replace('"',''))

top_ips = Counter(ips).most_common(3)
top_urls = Counter(urls).most_common(3)

print(top_ips)
print(top_urls)
```
