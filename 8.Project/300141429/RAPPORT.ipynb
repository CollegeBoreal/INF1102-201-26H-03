```python
import json
import re
from collections import Counter
import matplotlib.pyplot as plt

with open("data/articles.json") as f:
    titles = json.load(f)

words = []
for t in titles:
    words.extend(re.findall(r"\b[a-zA-Z]{4,}\b", t.lower()))

top = Counter(words).most_common(10)
top
```

```python
labels, values = zip(*top)

plt.figure()
plt.bar(labels, values)
plt.title("Mots les plus fréquents – BBC News")
plt.xlabel("Mots")
plt.ylabel("Fréquence")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Répartition des mots les plus fréquents")
plt.tight_layout()
plt.show()
```

```python
top5 = top[:5]
labels5, values5 = zip(*top5)

plt.figure()
plt.bar(labels5, values5)
plt.title("Top 5 mots les plus fréquents – BBC News")
plt.xlabel("Mots")
plt.ylabel("Fréquence")
plt.tight_layout()
plt.show()
```
