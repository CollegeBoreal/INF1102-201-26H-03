```python
## Introduction

Ce projet a pour objectif de surveiller la disponibilité de plusieurs sites web et de mesurer leur temps de réponse.

L’analyse est automatisée à l’aide :
- d’un script Bash pour lancer le programme
- d’un script Python pour analyser les données
- d’un rapport texte généré automatiquement

Cela permet de vérifier si un site est accessible et d’évaluer ses performances.
```

```python
import requests
import time
import matplotlib.pyplot as plt

sites = [
    "https://www.google.com",
    "https://www.github.com",
    "https://www.collegeboreal.ca"
]

temps = []
codes = []

for site in sites:
    try:
        start = time.time()
        response = requests.get(site)
        end = time.time()

        temps_reponse = (end - start) * 1000
        temps.append(temps_reponse)
        codes.append(response.status_code)

        print(site, "->", response.status_code, "-", round(temps_reponse, 2), "ms")

    except:
        temps.append(0)
        codes.append(0)
        print(site, "-> erreur")
```

```python
import requests
import time
import matplotlib.pyplot as plt

sites = [
    "https://www.google.com",
    "https://www.github.com",
    "https://www.collegeboreal.ca"
]

temps = []
codes = []

for site in sites:
    try:
        start = time.time()
        response = requests.get(site)
        end = time.time()

        temps_reponse = (end - start) * 1000
        temps.append(temps_reponse)
        codes.append(response.status_code)

        print(site, "->", response.status_code, "-", round(temps_reponse, 2), "ms")

    except:
        temps.append(0)
        codes.append(0)
        print(site, "-> erreur")
```

```python
plt.bar(sites, temps)
plt.title("Temps de réponse des sites web")
plt.xlabel("Sites")
plt.ylabel("Temps (ms)")
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure()
plt.plot(sites, temps, marker='o')
plt.title("Temps de réponse (ligne)")
plt.xlabel("Sites")
plt.ylabel("Temps (ms)")
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure()
labels = ["OK", "Erreur"]
valeurs = [codes.count(200), len(codes) - codes.count(200)]

plt.pie(valeurs, labels=labels, autopct='%1.1f%%')
plt.title("Répartition des statuts des sites")
plt.show()
```
