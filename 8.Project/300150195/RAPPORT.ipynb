## 🔤 Top 10 Mots Fréquents

````
import re
from collections import Counter

with open('output/rapport.txt') as f:
    lines = f.readlines()

mots = {}
for line in lines:
    match = re.match(r'(\w+)\s+█+\s+\((\d+)\)', line.strip())
    if match:
        mots[match.group(1)] = int(match.group(2))

for mot, count in sorted(mots.items(), key=lambda x: -x[1]):
    print(f'{mot:<20} {count}')
````
