## 1. Objectif
Vérifier la disponibilité de sites web et mesurer le temps de réponse.

```python
import requests
import time
import matplotlib.pyplot as plt

sites = [
    'https://www.google.com',
    'https://www.github.com',
    'https://www.wikipedia.org'
]

resultats = []
for site in sites:
    try:
        debut = time.time()
        r = requests.get(site, timeout=5)
        duree = round(time.time() - debut, 2)
        resultats.append({'site': site, 'status': r.status_code, 'temps': duree})
    except Exception as e:
        resultats.append({'site': site, 'status': 'ERREUR', 'temps': 0})

for r in resultats:
    print(f"{r['site']} - Status: {r['status']} - Temps: {r['temps']}s")

noms = [r['site'].replace('https://www.', '') for r in resultats]
temps = [r['temps'] for r in resultats]
couleurs = ['green' if r['status'] == 200 else 'red' for r in resultats]
statuts = [1 if r['status'] == 200 else 0 for r in resultats]

fig = plt.figure(figsize=(14, 10))
fig.suptitle('Monitoring des sites web', fontsize=16, fontweight='bold')

ax1 = fig.add_subplot(2, 2, 1)
ax1.bar(noms, temps, color=['green', 'blue', 'orange'])
ax1.set_title('Temps de réponse (s)')
ax1.set_xlabel('Sites')
ax1.set_ylabel('Temps (s)')

ax2 = fig.add_subplot(2, 2, 2)
ax2.bar(noms, [r['status'] if isinstance(r['status'], int) else 0 for r in resultats], color=couleurs)
ax2.set_title('Codes de statut HTTP')
ax2.set_xlabel('Sites')
ax2.set_ylabel('Code HTTP')

ax3 = fig.add_subplot(2, 2, 3)
dispo = sum(statuts)
non_dispo = len(statuts) - dispo
ax3.pie([dispo, non_dispo], labels=['Disponible', 'Indisponible'],
        colors=['green', 'red'], autopct='%1.0f%%', startangle=90)
ax3.set_title('Disponibilité globale')

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(noms, temps, marker='o', color='blue', linewidth=2, markersize=8)
ax4.fill_between(noms, temps, alpha=0.2, color='blue')
ax4.set_title('Courbe des temps de réponse')
ax4.set_xlabel('Sites')
ax4.set_ylabel('Temps (s)')

plt.tight_layout()
plt.savefig('output/figure.png', dpi=150)
plt.show()
print('✅ Figure sauvegardée')
```
