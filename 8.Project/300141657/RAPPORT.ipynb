# Rapport de monitoring site web
Projet 5 - Monitoring site web
Site surveille : https://www.google.com

```python
import pandas as pd
df = pd.read_csv("output/monitoring.csv")
df
```

```python
import matplotlib.pyplot as plt

df["response_time_ms"] = pd.to_numeric(df["response_time_ms"], errors="coerce")

plt.figure(figsize=(10,5))
plt.plot(df["timestamp"], df["response_time_ms"], marker="o")
plt.xticks(rotation=45, ha="right")
plt.title("Evolution du temps de reponse du site")
plt.xlabel("Date et heure")
plt.ylabel("Temps de reponse (ms)")
plt.tight_layout()
plt.show()
```
